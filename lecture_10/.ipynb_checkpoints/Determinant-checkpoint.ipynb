```python
#Final
MatrixA = [[1,2,3],
           [4,5,6],
           [7,8,9]]

MatrixC = [[4,3,6],
           [5,5,7],
           [8,9,4]]

MatrixD = [[234,2456,76],
           [345,645,43],
           [2346,4536, 234]]

MatrixE = [[2, 4, 1, 3],
           [0, -1, 2, 5],
           [3, 7, 2, 2],
           [1, 0, 4, -2]]


print(MatrixE)

def determinant(matrix):
    if len(matrix[0]) == 2: # base case: if matrix is 2x2 return determinant
        return matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0]
    else:
        det = 0 # initialize a variable that will store the sum of determinants we will calculate
        for i in range(len(matrix[0])): # loop through the first row
            subMatrix = [] # create a submatrix that will be used for the recursive call
            for j in range(len(matrix)-1): # loop through the rows minus the first one
                row = [] # initialize a row that will be put into the submatrix
                for k in range(len(matrix[0])): # loop through the columns of the j+1 row
                    if k != i: # if we are in the same column as i, don't add to submatrix
                        row += [matrix[j+1][k]] # add the element to the row
                subMatrix += [row] # add the row to the submatrix
            miniDet = determinant(subMatrix) #find the determinant of the submatrix
            det += ((-1) ** i) * matrix[0][i] * miniDet #calculate the determinant of for this loop and add to det
        return det # return det
    
print("determinant =", determinant(MatrixE))
```
